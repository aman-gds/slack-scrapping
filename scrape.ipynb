```python
import re
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time
import json
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException

# Function to clean the text by removing unwanted characters using regular expressions
def clean_text(text):
    # Remove unwanted characters and normalize the text
    cleaned_text = re.sub(r'[^\x00-\x7F]+', '', text)  # Remove non-ASCII characters
    cleaned_text = cleaned_text.replace('\n', ' ')  # Replace newline characters with spaces
    return cleaned_text

messages = []

# Find all message boxes
message_boxes = driver.find_elements(By.XPATH, '//div[contains(@class, "c-message_kit__message")]')

for message_box in message_boxes:
    try:
        message_data = {
            "message_text": clean_text(message_box.text),
            "replies": []
        }

        # Find the reply button within the message box (if available)
        try:
            reply_button = message_box.find_element(By.XPATH, './/button[@class="c-link--button c-message__reply_count"]')
        except NoSuchElementException:
            reply_button = None  # No reply button found

        if reply_button:
            ActionChains(driver).move_to_element(reply_button).click(reply_button).perform()
            time.sleep(3)  # Add a sleep to allow initial replies to load

            while True:
                # Re-locate reply boxes within the loop to avoid StaleElementReferenceException
                reply_boxes = driver.find_elements(By.XPATH, '//div[@class="c-message_kit__background c-message_kit__message c-message_kit__thread_message"]')
                for reply_box in reply_boxes:
                    message_data["replies"].append(clean_text(reply_box.text))

                # Simulate scrolling down with the keyboard
                for _ in range(5):  # You can adjust the number of times to press the down arrow key
                    ActionChains(driver).send_keys(Keys.ARROW_DOWN).perform()
                    time.sleep(3)  # Adjust the sleep time if needed

                # Check if there is a "Load more replies" button
                load_more_button = None
                try:
                    load_more_button = driver.find_element(By.XPATH, '//button[@class="c-button c-button--small c-message_kit__thread_loader__load_more c-button--outline c-button--small"]')
                except NoSuchElementException:
                    load_more_button = None

                if load_more_button:
                    # Click the "Load more replies" button
                    load_more_button.click()
                else:
                    break  # No more replies to load

            # Close the reply thread by clicking the "Back" button
            close_button = driver.find_element(By.XPATH, '//button[@class="c-button-unstyled c-icon_button c-icon_button--size_medium p-flexpane_header__control c-icon_button--default"]')
            close_button.click()

        # If there are no replies, add a message indicating that
        if not message_data["replies"]:
            message_data["replies"].append("There is no reply thread in this message")

        # Append the message data to the list
        messages.append(message_data)

    except StaleElementReferenceException:
        continue  # Handle stale element references and move on to the next message

# Create a dictionary to store the data
data = {
    "messages": messages
}

# Specify the JSON file name
json_file_name = "intros3.json"

# Save the data to a JSON file
with open(json_file_name, "w") as json_file:
    json.dump(data, json_file, indent=4, ensure_ascii=False)

print(f"Messages and replies have been saved to {json_file_name}")
```
